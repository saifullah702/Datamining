```python
plt.scatter(A,b , marker='*',color='blue')
plt.xlabel('ENGINESIZE')
plt.ylabel('CO2EMISSION')
plt.title('FuelConsumption')
```

```python
from sklearn.model_selection import train_test_split
A_train,A_test,b_train,b_test=train_test_split(A,b,test_size=20,random_state=1)
```

```python
from sklearn import tree
r=tree.DecisionTreeClassifier()
r.fit(A_train, b_train)

#predict
pred=r.predict(A_test)
```
